# Let's Plan

```python
df.drop(columns=['EmployeeCount','StandardHours','Over18'],inplace=True)
```

```python
data=df.copy()
data
```

```python
X = data.drop("Attrition", axis=1)
Y = data["Attrition"].map({'Yes': 1, 'No': 0})
```

```python
Y
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
X_train.info()
```

```python
# one hot encoding
trf1 = ColumnTransformer([
    ('ohe',OneHotEncoder(sparse=False,drop="first", handle_unknown='ignore'),[1,3,6,9,13,15,19])
],remainder='passthrough')
```

```python
dftrf1_X_test=pd.DataFrame(trf1.fit_transform(X_train))
```

```python
dftrf1_X_test
```

```python
dftrf1=trf1
dftrf1
```

```python

# Scaling
trf2 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,45))
])
```

```python
X_train
```

```python
# train the model
trf3 = DecisionTreeClassifier()
```

# Create Pipeline

```python
attr_pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)

])
```

```python
X_train
```

```python
# train
attr_pipe.fit(X_train,y_train)
```

# Explore the Pipeline

```python
# Code here
attr_pipe.named_steps
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# Predict
y_pred = attr_pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(attr_pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

# GridSearch using Pipeline

```python
# gridsearchcv
params = {
    'trf3__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(attr_pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```
